```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

# ticker_to_sector는 이미 존재한다고 가정
# ticker_to_sector = dict(zip(df["Yahoo_Ticker"], df["sector"]))

# 각 row에 sector 추가
df["sector"] = df["Ticker"].map(ticker_to_sector)

# 섹터별 그룹핑
sectors = df["sector"].dropna().unique()

with PdfPages("sector_MRL_plots.pdf") as pdf:
    for sector in sectors:
        df_sector = df[df["sector"] == sector]
        tickers_in_sector = df_sector["Ticker"].unique()

        plt.figure(figsize=(10, 6))
        for ticker in tickers_in_sector:
            temp = df_sector[df_sector["Ticker"] == ticker]
            plt.plot(temp["Threshold"], temp["MRL"], label=ticker)

        plt.title(f"MRL Curves - {sector}")
        plt.xlabel("Threshold")
        plt.ylabel("MRL")
        plt.legend(fontsize=6, ncol=2)  # 많은 경우 조절 가능
        plt.tight_layout()
        plt.show()
        pdf.savefig(fig, dpi=300)
        plt.close()
```
